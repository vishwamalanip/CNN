```python
data=[]
for i in categories:
    folder_path = os.path.join(mydir,i)
    
    if i=='cats':
        label = 0
    else:
        label= 1
    
    for j in os.listdir(folder_path):
        img_path =os.path.join(folder_path,j)
        img=cv2.imread(img_path)
        img=cv2.resize(img,(150,150))
        data.append([img,label])
    
```

```python
random.shuffle(data)
```

```python
x=[]
y=[]
for features,labels in data:
    x.append(features)
    y.append(labels)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train =x_train/255
x_test  =x_test/255
```

```python
#cnn
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
```

```python
# building the CNN model
model = Sequential()

model.add(Conv2D(64, kernel_size=(3,3), input_shape=(150,150,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
#for binaryclassification sigmoid #for regression linear  #for multiple classes softmax
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=6,validation_split=0.1)
```
